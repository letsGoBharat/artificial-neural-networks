```python
import numpy as np
import matplotlib.pyplot as plt

n = 100;
mA = [ 1.0, 0.5]; sigmaA = 0.5;
mB = [-1.0, 0.0]; sigmaB = 0.5;
mu = 0; #mean
sigma = 1; # standard deviation
np.random.normal(mu, sigma, 100)
A = np.array([np.random.normal(mu, sigma, 100) * sigmaA + mA[0],
              np.random.normal(mu, sigma, 100) * sigmaA + mA[1]]);
B = np.array([np.random.normal(mu, sigma, 100) * sigmaB + mB[0],
              np.random.normal(mu, sigma, 100) * sigmaB + mB[1]]);

plt.scatter(A[0, :],A[1, :]);
plt.scatter(B[0, :],B[1, :]);
plt.show();
```
